### 1. Buat graph berarah dengan networkx (minimal 9 node )

### 2. Hitung manual pagerank dari graph yang anda buat

#### Rumus


image.png

#### Iterasi 0

Kita inisialisasi terlebih dahulu, karena pada gambar diatas terdapat 9 node maka untuk iterasi 0 adalah `1/9` atau `0.111111`

#### Iterasi 1

#### Node P1

Pada P1 , hanya P4 yang mengarah pada P1. Sehingga nilai pada iterasi sebelumnya pada P4 akan di bagi dengan jumlah `outgoing links` pada P4

 **`PR(P1) = PR(P4)/Out(P4)`**


PR(P1)  = 0.111111 / 2 = 0.055556

#### Node P2

 **`PR(P2) = PR(P1)/Out(P1)`**


PR(P2) = 0.111111 / 1 = 0.111111

#### Node P3

 **`PR(P3) = PR(P2)/Out(P2)`**


PR(P3) = 0.111111 / 1 = 0.111111

#### Node P4


 **`PR(P4) = (PR(P3)/Out(P3)) + (PR(P8)/Out(P8)) `**


PR(P4) = (0.111111 / 1 ) + (0.111111 / 1) = 0.222222

#### Node P5

 **`PR(P5) = (PR(P4)/Out(P4)) + (PR(P9)/Out(P9)) `**


PR(P5) = (0.111111 / 2 ) + (0.111111 / 1) = 0.166667

#### Node P6

 **`PR(P6) = (PR(P5)/Out(P5))`**


PR(P6) = (0.111111 / 1 ) = 0.111111

#### Node P7

 **`PR(P7) = (PR(P6)/Out(P6))`**


PR(P7) = (0.111111 / 2 ) = 0.055556

#### Node P8

 **`PR(P8) = (PR(P6)/Out(P6)) + (PR(P7)/Out(P7))`**


PR(P8) = (0.111111 / 2 ) + (0.111111 / 2 ) = 0.111111

#### Node P9

 **`PR(P9) =  (PR(P7)/Out(P7))`**


PR(P9) = (0.111111 / 2 ) = 0.055556
