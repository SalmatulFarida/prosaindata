```python
import string 
import re
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from symspellpy import SymSpell ,Verbosity
import os
import requests
import numpy as np
```

```python
for i in data :
  data[data.index(i)] = i.split()
```

```python
#stopword
factory = StopWordRemoverFactory()
stopwords =  factory.create_stop_word_remover()
print(factory.get_stop_words())
```

```python
#prosesstopword
for i in data:
  for j in i:
    data[data.index(i)][i.index(j)] = stopwords.remove(j)
```

```python
#lowercase
punctuation = "!@#$%^&*()/\[]=_+<>?:.,;" 
for i in data:
  for j in i:
    data[data.index(i)][i.index(j)] = j.translate(str.maketrans("","",punctuation)).lower()
for i in data:
  for j in i:
    try:
      if(j[0] =='-'):
        data[data.index(i)][i.index(j)] = j.replace('-','')
    except:
      pass
#hapusnumber
for i in data:
  for j in i:
    data[data.index(i)][i.index(j)] = j.translate(str.maketrans("","",string.digits))
#hapuskarakterkosong
for i in data:
  for j in i:
    try:
      if(len(j) <=2):
        data[data.index(i)][i.index(j)] = data[data.index(i)].remove(j)
    except:
      pass
for i in data:    
  for j in range(i.count('')):
    i.remove('')
  for j in range(i.count(None)):
    i.remove(None)
```

```python
for i in data:    
  for j in range(i.count('-')):
    i.remove('-')
    for j in range(i.count(None)):
      i.remove(None)
punctuation = "!@#$%^&*()â€™_+<>?:.,;!@#$%^&*()/\[]=_" 
for i in data:
  for j in i:
    data[data.index(i)][i.index(j)] = j.translate(str.maketrans("","",punctuation)).lower()
for i in range(len(data)):
  data[i] = " ".join(data[i])
```

```python
#stemming
factory = StemmerFactory()





stemmer = factory.create_stemmer()
for i in data:
  data[data.index(i)] = stemmer.stem(i)
```

implmentasi TF-IDF

```python
X = data['Abstrak']
y = data['Kategori']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
vectorizer = TfidfVectorizer()
countvectorizer = CountVectorizer()
# hasil fit_transfrom
X_train_tfidf = vectorizer.fit_transform(X)
# menghitung jumlah
count_wm = countvectorizer.fit_transform(X)
# menampilkan hasil dan jumlah dari tfidf
count_tokens = countvectorizer.get_feature_names_out()
tfidf_tokens = vectorizer.get_feature_names_out()
df_countvect = pd.DataFrame(data = count_wm.toarray(),columns = count_tokens)
df_tfidfvect = pd.DataFrame(data = X_train_tfidf.toarray(),columns = tfidf_tokens)
print("Count Vectorizer\n")
print(df_countvect)
print("\nTD-IDF Vectorizer\n")
print(df_tfidfvect)
```
