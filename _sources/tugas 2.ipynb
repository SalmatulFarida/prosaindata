**Akses Database**

Untuk mengakses semua database dapat menggunakan Ngrok. di ngrok di setting TCP port dari masing-masing database. Agar Ngrok bisa berjalan bersamaan, maka di ngrok.yml di tambahkan sebuah tunnel yang mewakili port database. Untuk codenya ada dibawah ini:

```python
# This is formatted as code
tunnels:
  mysql:
    addr: 3306
    proto: tcp
  postgre:
    addr: 5432
    proto: tcp
```

lalu dapat mengakses database yang sesuai dengan TCP yang ada di dalam Ngrok. kemudian digantu untuk host dan portnya

** Get data dari Database**
